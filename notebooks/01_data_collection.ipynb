# 📈 Stock Price Predictor - Data Collection
### Fetching Apple Inc. (AAPL) historical stock data using yfinance

```python
# Importing the tools we need
import yfinance as yf      # fetches stock data from Yahoo Finance
import pandas as pd        # works with data like an Excel table
import os                  # helps us work with folders
```

```python
# Downloading Apple stock data from 2015 to 2024
# "AAPL" is Apple's ticker symbol on stock market
df = yf.download("AAPL", start="2015-01-01", end="2024-12-31")

# Show first 5 rows
df.head()
```

```python
# How many rows (days) and columns do we have?
print("Shape:", df.shape)

# What date range is covered?
print("Start date:", df.index.min())
print("End date:", df.index.max())

# Any missing values? (we need to know before cleaning)
print("\nMissing values:")
print(df.isnull().sum())
```

```python
# Save raw data to our data/raw folder
# "../" means go one folder back (from notebooks → Ai_Stock_Predictor)
os.makedirs("../data/raw", exist_ok=True)
df.to_csv("../data/raw/AAPL_raw.csv")

print("✅ Data saved to data/raw/AAPL_raw.csv")
print(f"Total trading days downloaded: {len(df)}")
```
